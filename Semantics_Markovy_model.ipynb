```python
import nltk
import random

song_lyrics_gujarati = [
    "જિંદગી એક સફર હૈ સુહાના, યહા ખુશીઓ કા પતા મિલતા હૈં।",
    "આજ કલ ઇશ્ક઼ માં જ઼િંદગી ગુજરાય છે, બેસ એક અરમાન દબાય છે।",
    "હવાઓ ની લપટ માં આનંદ લઈ, જાય છે મન ની ચિંતા દુર ગઈ।",
    "મારી મને એવી રહય છે મિઠાસ, એ મને જીવન ભર મેળવશે સાથ।",
    "અનન્ય પ્રેમમયી મોહબ્બત માં, મિલાય છે ખુશીઓ નો સાથ।",
    "મને જીવન ભર તેની જરૂરત છે, જે મને આનંદ નો મહસુસ કરાવે।",
    "જીવન માં આવી ગઈ ખુશી, મન માં છે મિઠું કરવું વાંચવું।",
    "પ્રેમ નો અંબર માં, મને વિશ્વાસ છે, મારી ખુશી અને મારી આશા।",
    "તમારી યાદો નો આવાજ, મને દુનિયા ની સૌથી મોટી ખુશી છે।",
    "મારા દિલ માં એવી ગાથા, જેને કોઈ જાને નહીં મિતાવું સાથ।",
    "જિંદગી ની રાહો માં, મને મીઠી યાદો નો મિલે સાથ।",
    "મન માં એવી આશા છે, જેને વ્યક્ત કરવા માટે મારી બોલો છે।",
    "તમે મારા પ્રેમ નો રાજ, મને મળે ખુશી નો હાજરા।",
    "તમે મારા દિલ નો ચહેરો, મારી જિંદગી નો અંતરાલો।",
    "તમે મારું જિંદગી નો મકસદ, મને મળે અનંત આનંદ।",
    "તમે મારા જીવન નો માલિક, મને મળે આનંદ નો હાજરા।",
    "મારા દિલ ની ગહરી જુબાન, મને મળે મોટી ખુશીઓ નો અનુભવ।",
    "મારો મન તમારું રાજ, મારી જિંદગી નો મોટું ખુશી નો સાથ।",
    "મારા દિલ નો સાથી, મારી જિંદગી નો રાજ।"
]

# tokenize the song lyrics
tokenized_lyrics_gujarati = [nltk.word_tokenize(lyric) for lyric in song_lyrics_gujarati]
```

```python
tokenized_lyrics_gujarati
```

```python

# created a transition model
markov_chain_gujarati = {}
for lyric in tokenized_lyrics_gujarati:
    for i in range(len(lyric) - 2):
        current_ngram = tuple(lyric[i:i + 2])
        next_word = lyric[i + 2]
        if current_ngram in markov_chain_gujarati:
            markov_chain_gujarati[current_ngram].append(next_word)
        else:
            markov_chain_gujarati[current_ngram] = [next_word]
```

```python
markov_chain_gujarati
```

```python
# Generate a large text based on the Gujarati song lyrics
current_ngram = random.choice(list(markov_chain_gujarati.keys()))
text = list(current_ngram)
length = 30
for _ in range(length - 2):
    if current_ngram in markov_chain_gujarati:
        next_word = random.choice(markov_chain_gujarati[current_ngram])
        text.append(next_word)
        current_ngram = tuple(text[-2:])
    else:
        break
```

```python
generated_large_text = ' '.join(text)
print(generated_large_text)
```
